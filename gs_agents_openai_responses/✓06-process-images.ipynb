```python
import os
from dotenv import load_dotenv
from rich import print

from semantic_kernel.agents import AgentThread, OpenAIResponsesAgent
from semantic_kernel.contents import (
    TextContent,
    ImageContent,
    AuthorRole,
    ChatMessageContent,
)
```

```python
load_dotenv("../.env/bailian", override=True)
print(os.getenv("OPENAI_RESPONSES_MODEL_ID"))
```

```python
from pathlib import Path

image_path = "../python-samples/getting_started_with_agents/resources/cat.jpg"
image_path = Path(image_path).absolute().as_posix()
```

```python
client = OpenAIResponsesAgent.create_client()
agent_v = OpenAIResponsesAgent(
    ai_model_id=os.getenv("OPENAI_RESPONSES_MODEL_ID"),
    client=client,
    instructions="Answer questions about the provided images.",
    name="VisionAgent",
    store_enabled=False,
)
```

指定的URI是先在本地下载完成，转换成BASE64字符串后才传到base_url去。

```python
user_messages = [
    ChatMessageContent(
        role=AuthorRole.USER,
        items=[
            TextContent(text="Describe this image."),
            ImageContent(
                # uri="https://upload.wikimedia.org/wikipedia/commons/thumb/4/47/New_york_times_square-terabass.jpg/1200px-New_york_times_square-terabass.jpg"
                uri="https://img1.qunarzz.com/travel/d9/1612/89/8d880333600a0cb5.jpg_r_640x426x70_e744d368.jpg"
            ),
        ],
    ),
    ChatMessageContent(
        role=AuthorRole.USER,
        items=[
            TextContent(text="What is the main color in this image?"),
            ImageContent(
                # uri="https://upload.wikimedia.org/wikipedia/commons/5/56/White_shark.jpg"
                uri="https://science.haichangoceanpark.com/wp-content/uploads/2023/04/bai-jing-1.jpg"
            ),
        ],
    ),
    ChatMessageContent(
        role=AuthorRole.USER,
        items=[
            TextContent(text="Is there an animal in this image?"),
            ImageContent.from_image_file(image_path),
        ],
    ),
]
```

```python
thread: AgentThread = None
for user_input in user_messages:
    print(f"[red]User ▶︎ {str(user_input)}[/red]")

    response = await agent_v.get_response(messages=user_input, thread=thread)
    print(f"[green]Agent ▶︎ {response.content}[/green]")

    thread = response.thread
```
