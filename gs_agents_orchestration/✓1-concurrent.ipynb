# 1 并发调度 ConcurrentOrchestration

```python
from rich import print
from IPython.display import display, Markdown, Latex
from dotenv import load_dotenv
```

```python
from semantic_kernel.agents import ChatCompletionAgent, ConcurrentOrchestration
from semantic_kernel.connectors.ai.open_ai import OpenAIChatCompletion
```

```python
load_dotenv("../.env/qiniu", override=True)
```

```python
physics_agent = ChatCompletionAgent(
    name="PhysicsExpert",
    instructions="You are an expert in physics. You answer questions from a physics perspective.",
    service=OpenAIChatCompletion(),
)
chemistry_agent = ChatCompletionAgent(
    name="ChemistryExpert",
    instructions="You are an expert in chemistry. You answer questions from a chemistry perspective.",
    service=OpenAIChatCompletion(),
)
```

```python
ccor = ConcurrentOrchestration(members=[physics_agent, chemistry_agent])
```

```python
from semantic_kernel.agents.runtime import InProcessRuntime

runtime = InProcessRuntime()
runtime.start()
```

```python
orchestration_result = await ccor.invoke(
    task="What is temperature?",
    runtime=runtime,
)
value = await orchestration_result.get(timeout=320)
```

```python
for item in value:
    print(f"[#2ECC71]{item.name} ▶︎ [/#2ECC71]")
    display(Markdown(Latex(item.content).data))
```

```python
await runtime.stop_when_idle()
```

# 1a: 1 + 结构化输出

```python
from semantic_kernel.agents.orchestration.tools import structured_outputs_transform
from pathlib import Path
```

```python
# load_dotenv("../.env/firework_ai", override=True)
load_dotenv("../.env/bailian", override=True)
# 都能正常生成结构化结果
```

```python
from pydantic import BaseModel


class ArticleAnalysis(BaseModel):
    """A model to hold the analysis of an article."""

    themes: list[str]
    sentiments: list[str]
    entities: list[str]
```

```python
theme_agent = ChatCompletionAgent(
    name="ThemeAgent",
    instructions="You are an expert in identifying themes in articles. Given an article, identify the main themes.",
    service=OpenAIChatCompletion(),
)
sentiment_agent = ChatCompletionAgent(
    name="SentimentAgent",
    instructions="You are an expert in sentiment analysis. Given an article, identify the sentiment.",
    service=OpenAIChatCompletion(),
)
entity_agent = ChatCompletionAgent(
    name="EntityAgent",
    instructions="You are an expert in entity recognition. Given an article, extract the entities.",
    service=OpenAIChatCompletion(),
)
```

```python
ccor2 = ConcurrentOrchestration[str, ArticleAnalysis](
    members=[theme_agent, sentiment_agent, entity_agent],
    output_transform=structured_outputs_transform(
        ArticleAnalysis, OpenAIChatCompletion()
    ),
)
```

```python
file_path = "../python-samples/getting_started_with_agents/resources/Hamlet_full_play_summary.txt"
task = Path(file_path).open().read()
```

```python
runtime = InProcessRuntime()
runtime.start()
```

```python
orchestration_result = await ccor2.invoke(
    task=task,
    runtime=runtime,
)
value = await orchestration_result.get(timeout=320)
```

```python
if isinstance(value, ArticleAnalysis):
    print(value.model_dump_json(indent=2))
else:
    print("Unexpected result type:", type(value))
```

```python
await runtime.stop_when_idle()
```
